**Created funtion to remove stop words**

```python
def remove_stop_words(stop_words,path_test_file):
  filename = path_test_file
  file = open(filename, 'rt')
  text = file.read()
  tokens = word_tokenize(text)
  #tokens=tokens.lower()
  global remove_words
  remove_words = [word for word in tokens if word not in stop_words] ### comprehension function to remove stop words from the text file
  remove_words = [word for word in remove_words if word.isalpha()] ## comprehension function to display only alphabets in the text file
  #print("List without stop words:",remove_words[:300])
  #print("tokens:",tokens)
  #print("length of words:",len(remove_words))
  positive_words(remove_words)
  Negative_words(remove_words)
  uncertainty_words(remove_words)
  constraing_words(remove_words)
```

**Positive Score**

```python
def positive_words(word1):
  list1 =[]
  positive_path ="/content/Positive.txt"
  words = open(positive_path,"rt")
  words= words.read()
  pwords = words.split()
  global positive_count
  positive_count = 0
  for i in word1:
    if i in pwords:
      positive_count=positive_count+1
    else:
      continue
  #list1.append(positive_count)
  #print(list1)
  print("positive score:",positive_count)
```

**Negative Score**

```python
def Negative_words(word1):
  positive_path ="/content/Negative.txt"
  words = open(positive_path,"rt")
  words= words.read()
  pwords = words.split()
  global negative_count
  negative_count = 0
  for i in word1:
    if i in pwords:
      negative_count=negative_count - 1
    else:
      continue
  negative_count = -negative_count
  print("Negative score:",negative_count)
```

**Polarity Score**

```python
def polarity_calculation():
  global polarity_score
  if (positive_count !=0 or negative_count!=0):
    polarity_score = (positive_count-negative_count)/(positive_count+negative_count)
    polarity_score = polarity_score+0.000001
    print("Polarity score:",round(polarity_score,2))
  else:
    print("Polarity score:",0)
```

**Additional calculations for reference**

```python
def subjectivity_calculation():
  subject_score=(positive_count+negative_count)/len(remove_words)
  subject_score = subject_score+0.000001
  print(subject_score)
```

```python
def sentiment_score():
  if (polarity_score < -0.5):
    print("Most Negative")
  elif (polarity_score < 0  and polarity_score > -0.5):
    print("Negative")
  elif (polarity_score==0):
    print("Neutral")
  elif (polarity_score > 0 and polarity_score < 0.5):
    print("Positive")
  elif (polarity_score > 0.5):
    print("Very Positive")
```

**Percentage of complex words**

```python
def percent_complex_words():
  global ccount
  global percentage_complex
  ccount = 0
  for word in remove_words:
    vcount = 0
    if (word.endswith(("es","ed"))):
      continue
    else:
      for i in word:
        if (i=="a" or i=="e" or i =="i" or i=="o" or i=="u"):
          vcount = vcount + 1
        if (vcount > 2):
          ccount = ccount+1
  if (len(remove_words) is not 0):
    percentage_complex = ccount/len(remove_words)
    print("Percentage of complex words: ",round(percentage_complex,2))
  else:
    print("Percentage of complex words: ",0)
```

**Uncertainty score calculation**

```python
def uncertainty_words(word1):
  positive_path ="/content/uncertainty.txt"
  words = open(positive_path,"rt")
  words= words.read()
  uwords = words.split()
  global uncertainty_count
  uncertainty_count = 0
  for i in word1:
    if i in uwords:
      uncertainty_count=uncertainty_count+1
    else:
      continue
  print("Uncertainty score:",uncertainty_count)
```

**Constraining score calculation**

```python
def constraing_words(word1):
  constrain_path ="/content/constraining.txt"
  words = open(constrain_path,"rt")
  words= words.read()
  cwords = words.split()
  global constrain_words
  constrain_words = 0
  for i in word1:
    if i in cwords:
      constrain_words=constrain_words + 1
    else:
      continue
  print("Constraing score:",constrain_words)
```

**Calculations for Positive, Negative, Uncertainty and Constrain Proportions**

```python
def proportion_count():
  if (len(remove_words)!=0):
    global positive_proportion, negative_proportion,uncertainty_proportion,constrain_proportion
    positive_proportion = positive_count/len(remove_words)
    negative_proportion = negative_count/len(remove_words)
    uncertainty_proportion = uncertainty_count/len(remove_words)
    constrain_proportion = constrain_words/len(remove_words)
    print("Positive proportion: ",round(positive_proportion,0))
    print("Negative proportion: ",round(negative_proportion,0))
    print("Uncertainty proportion: ",round(uncertainty_proportion,0))
    print("constrain proportion: ",round(constrain_proportion,0))
```
